```python
import numpy as np
import pandas as pd
import math
import quandl
from sklearn import preprocessing,model_selection,svm
from sklearn.linear_model import LinearRegression
df=quandl.get('WIKI/YHOO',authtoken='h5CVihsz4vubHZf5Apqz')
df.head(10)
```

```python
df=df[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume']]
df.head()
```

```python
df['HL_PCT']=(df['Adj. High']-df['Adj. Low'])/df['Adj. Close']*100.0
df.head()
```

```python
df['PCT_change']=(df['Adj. Close']-df['Adj. Open'])/df['Adj. Close']*100.0
df.head()
```

```python
df=df[['Adj. Close','Adj. Volume','HL_PCT','PCT_change']]
print(df.head())
```

```python
forecast_col='Adj. Close'
df.fillna(value=-99999,inplace=True)
forecast_out=int(math.ceil(0.01*len(df)))
df.head()
```

```python
df['label']=df[forecast_col].shift(-forecast_out)
df.head()
```

```python
df.dropna(inplace=True)
x=np.array(df.drop(['label'],1))
y=np.array(df['label'])
x=preprocessing.scale(x)
y=np.array(df['label'])

```

```python
x_train, x_test, y_train, y_test =model_selection.train_test_split(x, y, test_size=0.2)
clf=svm.SVR()
clf.fit(x_train, y_train)
confidence = clf.score(x_test, y_test)
print(confidence)
```

```python
import numpy as np
from sklearn import preprocessing, model_selection, neighbors
import pandas as pd
df = pd.read_csv('breast-cancer-wisconsin.txt')
df.replace('?',-99999, inplace=True)
df.drop(['id'], 1, inplace=True)
X = np.array(df.drop(['class'], 1))
y = np.array(df['class'])
X_train, X_test, y_train, y_test =model_selection.train_test_split(X, y, test_size=0.2)
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train, y_train)
confidence = clf.score(X_test, y_test)
print(confidence)
```

```python
example_measures = np.array([[4,2,1,1,1,2,3,2,1]])
example_measures.reshape(len(example_measures), -1)
prediction = clf.predict(example_measures)
print(prediction)
```

```python
import numpy as np
from sklearn import preprocessing , model_selection,neighbors, svm
import pandas as pd
df = pd.read_csv('breast-cancer-wisconsin.txt')
df.replace('?',-99999, inplace=True)
df.drop(['id'], 1, inplace=True)
X = np.array(df.drop(['class'], 1))
y = np.array(df['class'])
X_train, X_test, y_train, y_test =model_selection.train_test_split(X, y, test_size=0.2)
clf = svm.SVC()
clf.fit(X_train, y_train)
confidence = clf.score(X_test, y_test)
print(confidence)
example_measures = np.array([[4,2,1,1,1,2,3,2,1]])

example_measures.reshape(len(example_measures), -1)
prediction = clf.predict(example_measures)
print(prediction)
```
